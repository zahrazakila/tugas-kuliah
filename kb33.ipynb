```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import matplotlib as mpl
```

```python
df = pd.read_csv("data.csv")

reg = LinearRegression()
reg.fit(df[['x1','x2']].values,df.y)
```

```python
mpl.rcParams['legend.fontsize'] = 12
fig = plt.figure()
ax = fig.add_subplot(projection ='3d')
ax.scatter(df.x1, df.x2, df.y)
ax.legend()
ax.view_init(45, 0)
plt.show()
```

```python
b = reg.coef_
a = reg.intercept_

print('Intercept = ', a)
print('Koefisien regresi = ', b[0],'dan', b[1])
```

```python
X_baru = np.array([[8,125]]).reshape(1,2)

prediksi_Y = reg.predict(X_baru)
print('Hasil prediksi untuk x1=8 dan x2=125 yaitu', prediksi_Y)
```
